from random import randint
from time import sleep

def get_playlist_tracks(playlist_id):
    results = sp.user_playlist_tracks("spotify",playlist_id)
    tracks = results['items']
    while results['next']!=None:
        results = sp.next(results)
        tracks = tracks + results['items']
        sleep(randint(1,3))
    return tracks

# this will take at least around num_pages_in_playlist * (avg_sleep_time + processing_time) = 53 * (2+0.1) = 110 seconds

##### Managing imbalance in the dataset

###### Downsampling

```python
df_group_1_test['popularity'].value_counts()
```

```python
from sklearn.utils import resample
```

```python
category_0 = df_group_1_test[df_group_1_test['popularity'] == 0]
category_1 = df_group_1_test[df_group_1_test['popularity'] == 1]
```

```python
category_0_undersampled = resample(category_0, #<- downsample from here
                                    replace=False, #<- no need to reuse data now, we have an abundance
                                    n_samples = len(category_1),
                                    random_state=0)
```

```python
print(category_0_undersampled.shape)
print(category_1.shape)
```

```python
data_downsampled = pd.concat([category_0_undersampled, category_1], axis=0)
data_downsampled = data_downsampled.sample(frac =1) #randomize the rows
```

```python
data_downsampled['popularity'].value_counts()
```

```python
X = data_downsampled.drop(['popularity','genre','artist_name','track_name','track_id'], axis = 1)
y = data_downsampled['popularity']
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
X
```

```python
y
```

```python
X_train_scaled = transformer.transform(X_train)
X_test_scaled = transformer.transform(X_test)
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_scaled, y_train)
```

```python
print('accuracy:',LR.score(X_train_scaled, y_train))
# d_accuracy = LR.score(X_train_scaled, y_train)
```

```python
y_pred = LR.predict(X_test_scaled)
confusion_matrix(y_test, y_pred)
```

###### Upsampling

```python
category_1_upsampled = resample(category_1, #<- sample from here
                                    replace=True, #<- we need replacement, since we don't have enough data otherwise
                                    n_samples = len(category_0),#<- make both sets the same size
                                    random_state=0)
```

```python
print(category_1_upsampled.shape)
print(category_0.shape)
```

```python
data_upsampled = pd.concat([category_1_upsampled, category_0], axis=0)
data_upsampled = data_upsampled.sample(frac =1) #randomize the rows
```

```python
data_upsampled['popularity'].value_counts()
```

```python
X = data_upsampled.drop(['popularity','genre','artist_name','track_name','track_id'], axis = 1)
y = data_upsampled['popularity']
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
X_train_scaled = transformer.transform(X_train)
X_test_scaled = transformer.transform(X_test)
```

```python
LR = LogisticRegression(random_state=0, solver='lbfgs')
LR.fit(X_train_scaled, y_train)
```

```python
print('accuracy:',LR.score(X_train_scaled, y_train))
# d_accuracy = LR.score(X_train_scaled, y_train)
```

```python
y_pred = LR.predict(X_test_scaled)
confusion_matrix(y_test, y_pred)
```
